Question 2: Which videos have the highest rate of views per day?

```python
df2 = pd.read_csv('datavis_videos_q2.csv')
df2
```

We created a new column that takes the difference of the total views and the average total views, then stored it in a new column called "View Difference".

```python
df2['View Difference'] = df2['Total Views'] - df2['Average Total Views']
df2
```

We created a dropdown selection box for 3 months, and displayed the view difference with respect to the day. This is just data for lectures 12-21.

```python
@interact(month=['September','October','November'])
def f(month):
    df3 = df2[df2['Publish Month'] == month]
videos = ['COMP 4304 / 6934 Lecture 12', 'COMP 4304 / 6934 Lecture 13','COMP 4304 / 6934 Lecture 14',
          'COMP 4304 / 6934 Lecture 15','COMP 4304 / 6934 Lecture 16','COMP 4304 / 6934 Lecture 17',
         'COMP 4304 / 6934 Lecture 18','COMP 4304 / 6934 Lecture 19','COMP 4304 / 6934 Lecture 20',
          'COMP 4304 / 6934 Lecture 21',]
df3 = df2[df2['Video title'].isin(videos)]
px.line(x='Day', y='View Difference', data_frame = df3, color = 'Video title',range_x=[0,70], range_y=[-60,60])    
```

Question 3: Which day of the week is most popular for watching videos?

```python
df4 = pd.read_csv('datavis_videos_q3.csv')
```

Create a new dataframe that is grouped by the days of the week. This dataframe takes the average daily views, and displays
it with respect to the specific day of the week. Then, I reindexed the dataframe so Monday was on top, and the days were in order going clock-wise.

```python
df4 = df4[['nameofday','Total Daily Views']].groupby('nameofday').mean().reset_index().sort_values('nameofday')
df4 = df4.reindex([1,5,6,4,0,2,3])
```

```python
px.line_polar(r='Total Daily Views', theta='nameofday',data_frame = df4, line_shape='spline',
              line_close = True)
```
