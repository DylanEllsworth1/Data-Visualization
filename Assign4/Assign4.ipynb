Import squarify library, and make a treemap of all 3 game companies on the same visualization.

```python
import squarify
games = pd.read_csv('game_consoles.csv')
games
```

```python
sum_units = games['Units_sold'].sum()
width = 35001
height = 44125
sum_units / 35001
```

```python
Sony = games[games[('Company')]=='Sony']
Microsoft = games[games[('Company')]=='Microsoft']
Nintendo = games[games[('Company')]=='Nintendo']
```

```python
fig = plt.figure(figsize=(9,4))
ax1 = fig.add_axes([0.9,0.5,1,0.5])
ax2 = fig.add_axes([0.9,0.0,1,0.5])
ax3 = fig.add_axes([0.0,0.0,0.9,1])
values1 = squarify.normalize_sizes(Sony.Units_sold,35001,44125)
values2 = squarify.normalize_sizes(Microsoft.Units_sold,35001,44125)
values3 = squarify.normalize_sizes(Nintendo.Units_sold,35001,44125)
squarify.plot(values1,label=Sony.Console,color='blue',pad=True,ax=ax1)
squarify.plot(values2,label=Microsoft.Console,color='red',pad=True,ax=ax2)
squarify.plot(values3,label=Nintendo.Console,color='green',pad=True,ax=ax3)
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')

plt.title('Console Sales')
plt.show()
```
